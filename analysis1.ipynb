```python
#impute missing review rating by median (category wise)
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
category_wise_stats = df.groupby("Category")["Review Rating"].agg(["mean", "median"]).round(2)
print(category_wise_stats)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
bins = [17, 25, 40, 55, 70]   # 17 ensures 18 is included
labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"] = pd.cut(df["age"], bins=bins, labels=labels)
df[["age", "age_group"]].head(10)
```

```python
#define purchase_frequency_days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df['purchase_frequency_days'] = df['purchase_frequency_days']
df[["frequency_of_purchases", "purchase_frequency_days"]].head()
```

```python
df = df.drop('promo_code_used', axis=1) 
#this is because promo cod eused and discount applied have the same values - so removing redundancy
```

```python
df.columns
```

```python
#age group wise shipping method
shipping_dist = (
    df.groupby(["age_group", "shipping_type"])
      .size()
      .reset_index(name="Count")
)

print(shipping_dist)
```

```python
shipping_counts = df.groupby(["age_group", "shipping_type"]).size()

idx = shipping_counts.groupby(level=0).idxmax()

preferred_shipping = shipping_counts.loc[idx].reset_index(name="Count")

print(preferred_shipping)
```
